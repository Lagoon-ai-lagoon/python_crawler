練習題：計算體重指數 BMI
BMI（Body Mass Index，身體質量指數）是評估體重是否健康的常用指標。

計算公式：
BMI = 體重(公斤) ÷ 身高(公尺)²

題目要求：

建立變數 weight 儲存體重 72 公斤
建立變數 height 儲存身高 175 公分（注意：公式需要公尺，請先除以 100）
使用 ** 運算子計算身高的平方
計算 BMI 值並儲存在變數 bmi 中
使用 round() 函式將 BMI 四捨五入到小數點後 2 位
印出結果
參考輸出：
23.35

提示：

身高需轉換成公尺：height = 175 / 100
平方使用 ** 運算子：height ** 2
四捨五入：round(bmi, 2)

```python
weight = 72
height = 175

#height = (height/100)
#height = height**2

#print(weight/height)

#bmi=round(weight/height,2)

#print(bmi)

bmi=round((weight/(height/100)**2),2)

print(bmi)

```
